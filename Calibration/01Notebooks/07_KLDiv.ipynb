```python
from Functions.databuilder import dataBuilder
from Functions.data_preparation_for_ut import data_preparation_for_ut
from Functions.ut_plots import ut_plot

from scipy.special import kl_div,softmax


import warnings
warnings.filterwarnings('ignore')
```

```python
data1 = dataBuilder('293532')
data2 = dataBuilder('307328')
data3 = dataBuilder('307326')
data4 = dataBuilder('307327')
data5 = dataBuilder('293533')
```

```python
#Preprocessing Data
#Softmax has Axis=0, thus Normalising Per Stave
def datasetbuilding_for_kl(data):
    data = data[['Sector','Staves','Rows','PedestalValue']]
    data = data[data['Sector'].str.startswith('UTaX')]

    dataset = softmax(data_preparation_for_ut(data).fillna(0))

    return dataset
```

```python
import os
path_ = './Plots/07KL_Div'
os.makedirs(path_,exist_ok=True)
```

```python
#KLDiv Baseline with Data 2
ut_plot(kl_div(datasetbuilding_for_kl(data1),datasetbuilding_for_kl(data2)),show_fig=True,path=path_,filename='Data2',save_fig=True,vmin_vmax_scaled=True)
```

```python
#KLDiv Baseline with Data 3
ut_plot(kl_div(datasetbuilding_for_kl(data1),datasetbuilding_for_kl(data3)),show_fig=True,path=path_,filename='Data3',save_fig=True,vmin_vmax_scaled=True)
```

```python
#KLDiv Baseline with Data 4
ut_plot(kl_div(datasetbuilding_for_kl(data1),datasetbuilding_for_kl(data4)),show_fig=True,path=path_,filename='Data4',save_fig=True,vmin_vmax_scaled=True)
```

```python
#KLDiv Baseline with Data 5
ut_plot(kl_div(datasetbuilding_for_kl(data1),datasetbuilding_for_kl(data5)),show_fig=True,path=path_,filename='Data5',save_fig=True,vmin_vmax_scaled=True)
```
