```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

from Functions.databuilder import dataBuilder

import warnings
warnings.filterwarnings('ignore')
sns.set(rc = {'figure.figsize':(15, 5)})
```

```python
data1 = dataBuilder(293528)
data2 = dataBuilder(298595)
data3 = dataBuilder(306325)
```

```python
data1.shape
```

```python
#EDA
import os
path = './Plots/01EDANotebook'
os.makedirs(path,exist_ok=True)
```

```python
sns.histplot(data1['PedestalValue'],color='tab:red',alpha=0.5,bins=50,label=data1['RunNumber'][0])
sns.histplot(data2['PedestalValue'],color='tab:blue',alpha=0.5,bins=50,label=data2['RunNumber'][0])
sns.histplot(data3['PedestalValue'],color='tab:green',alpha=0.5,bins=50,label=data3['RunNumber'][0])
plt.savefig(f'{path}/Histogram.png')
plt.title('Histograms for Sample Runs')
plt.legend(title="Run IDs")
plt.show()
```

```python
plt.subplot(1, 3, 1)
sns.histplot(data=data1['PedestalValue'],color='tab:red',bins=25)
plt.title(data1['RunNumber'][0])

plt.subplot(1, 3, 2)
sns.histplot(data=data2['PedestalValue'],color='tab:blue',bins=25)
plt.title(data2['RunNumber'][0])

plt.subplot(1, 3, 3)
sns.histplot(data=data3['PedestalValue'],color='tab:green',bins=25)
plt.title(data3['RunNumber'][0])

plt.tight_layout()
plt.savefig(f'{path}/SplitHistograms.png')

plt.show()
```

```python
len=536576
run_list = [293528,298595,306325]
for i, run_id in enumerate(run_list):
    # Call dataBuilder and get the data
    df = dataBuilder(run_id)
    
    # Generate x-values based on the index and length
    x_values = np.arange(i * len, (i + 1) * len)
    
    # Plot the data
    sns.lineplot(x=x_values, y=df["PedestalValue"], label=f"Run {run_id}")

# Finalize the plot
plt.legend(title="Run IDs")
plt.title("Pedestal Value Across Runs")
plt.xlabel("Index")
plt.ylabel("Pedestal Value")
plt.savefig(f'{path}/PedestalValues.png')
plt.show()
```
